```python
import pandas as pd
from arboreto.utils import load_tf_names
from arboreto.algo import grnboost2
if __name__ == '__main__':
 # ex_matrix is a Data Frame with gene names as column names
 ex_matrix = pd.read_csv('NeuroblastsWT.txt', sep=',')
 # tf_names are read using a utility function included in Arboreto
 tf_names = load_tf_names('TFs.txt')
 print(tf_names[:5])
 network = grnboost2(expression_data=ex_matrix,
 tf_names=tf_names)
 network.to_csv('[ctype].tsv', sep='\t', index=False, header=False)#Red de output
 print("Done")
```
